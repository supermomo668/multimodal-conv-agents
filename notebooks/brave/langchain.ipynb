```python
import os
import click
import requests
import json 
import warnings
from typing import *

from dotenv import load_dotenv
from langchain_community.tools import BraveSearch
from langchain_community.document_loaders import PlaywrightURLLoader

from datetime import datetime 

os.chdir("../../")
print(os.getcwd())
# Load environment variables from .env file
load_dotenv()
```

```python
from langchain_community.document_loaders import WebBaseLoader

async def load_urls_playwright(urls: List[str]=[
    "https://www.youtube.com/watch?v=dQw4w9WgXcQ",
    "https://goo.gl/maps/NDSHwePEyaHMFGwh8",
  ]):
  loader = PlaywrightURLLoader(urls=urls, remove_selectors=["header", "footer"])
  return  await loader.aload()
  
# docs = await load_urls_playwright()
WebBaseLoader("https://www.youtube.com/watch?v=dQw4w9WgXcQ").load()
```

```python
def basic_tool():
  return BraveSearch.from_api_key(
    api_key=os.getenv("BRAVE_API_KEY"), 
    search_kwargs={"count": 3}
  )

def advanced_url_extractor(query: str):
  tool = basic_tool()
  # load json_str
  urls = [l.get("link") for l in json.loads(tool.run(query))]
  content = load_urls_playwright(urls)
  return content
content = await advanced_url_extractor("Last Fed Meeting")
```
