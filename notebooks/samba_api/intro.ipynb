```python
def call_sambaverse_api(
  input_prompt, 
  system_prompt="Respond for marketing uses only.", 
  model_name="Meta/llama-2-13b-chat-hf", 
  max_tokens=1024, 
  temperature=0.7, 
  top_k=50, 
  top_p=0.95
  ):
    # Get API key from environment variable
    api_key = os.getenv('SAMBA_API_KEY')
    if not api_key:
        raise ValueError("API key not found. Please set the SAMBA_API_KEY environment variable.")
    
    # Define the API endpoint
    api_url = 'https://sambaverse.sambanova.net/api/predict'
    
    # Construct the headers
    headers = {
        'Content-Type': 'application/json',
        'key': api_key,
        'modelName': model_name
    }
    
    # Construct the payload
    payload = {
        "instance": json.dumps({
            "conversation_id": "sambaverse-conversation-id",
            "messages": [
                {"message_id": 0, "role": "system", "content": system_prompt},
                {"message_id": 1, "role": "user", "content": input_prompt}
            ]
        }),
        "params": {
            "do_sample": {"type": "bool", "value": True},
            "max_tokens_to_generate": {"type": "int", "value": max_tokens},
            "process_prompt": {"type": "bool", "value": True},
            "repetition_penalty": {"type": "float", "value": 1.0},
            "return_token_count_only": {"type": "bool", "value": False},
            "select_expert": {"type": "str", "value": model_name},
            "stop_sequences": {"type": "str", "value": ""},
            "temperature": {"type": "float", "value": temperature},
            "top_k": {"type": "int", "value": top_k},
            "top_p": {"type": "float", "value": top_p}
        }
    }
    
    # Make the POST request
    try:
        response = requests.post(
            api_url, headers=headers, 
            data=json.dumps(payload)
        )
        # Check for rate limit exceeded
        if response.status_code == 429:
            warnings.warn("Rate limit exceeded. Please try again later.", UserWarning)
        
        # Raise an error for other non-success status codes
        response.raise_for_status()
        
        # Return the response content
        return response.json()
    except Exception as e:
        return {
            "error": e,
            "response": response.text
        }
```
